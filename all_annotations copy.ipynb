## Main

```python
import os
import pandas as pd # We use pandas to manipulate the data
import json  #Load the the data in json

# Automatically iterate through all annotations files and merge
folder_path = "path to load all annotations"
all_annotations = []
for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)
    if os.path.splitext(file_path)[1] == ".json":
        results = json.load(open(file_path))
        examples= results['examples']
        annotations = sum(map(lambda x:x['annotations'],examples),[])
        for x in annotations:
            all_annotations.append(x)

print(len(all_annotations))
```

```python
# save merged annotations as json file
with open('all_annotations.json', 'w') as f:
    json.dump(all_annotations, f)
```
